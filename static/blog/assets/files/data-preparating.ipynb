## 1.Smoothing Equi-width binning and Equi-depth binning


### Equi-width binning

equi-width binning may consider as fair cutoff by the range of data range


```python
df_eq_bin = df_copy.copy()
bin_label = ['bin_{}'.format(x) for x in range(8)]  # type: ignore
df_eq_bin['equ_width__departure_delay_in_minutes'] = pandas.cut(
    df_eq_bin['Departure Delay in Minutes'],
    8, labels=bin_label
)
df_eq_bin['equ_width__arrival_delay_in_minutes'] = pandas.cut(
    df_eq_bin['Arrival Delay in Minutes'],
    8, labels=bin_label
)
```

```python
# for display after binning

seaborn.histplot(data=df_eq_bin, x="equ_width__departure_delay_in_minutes",
                 label="Departure Delay in Minutes", kde=True,
                 cbar_kws={
                     "color": "skyblue", "alpha": 0.3
                 })
plt.legend()
plt.show()

seaborn.histplot(data=df_eq_bin, x="equ_width__arrival_delay_in_minutes",
                 label="Arrival Delay in Minutes", kde=True,
                 cbar_kws={
                     "color": "red", "alpha": 0.3, "pad": 0.3
                 })

plt.legend()
plt.show()
```

### Equi-depth binning

Equi-depth binning are based on the frequency of range group


```python

df_eq_bin['equ_depth__departure_delay_in_minutes'] = pandas.qcut(
    df_eq_bin['Departure Delay in Minutes'],
    8, duplicates='drop'
)
grouplist = df_eq_bin['equ_depth__departure_delay_in_minutes'].unique(
).tolist()
grouplist.sort()
group_mapping = {g: "bin_{}".format(i) for i, g in enumerate(grouplist)}
df_eq_bin['equ_depth__departure_delay_in_minutes'] = df_eq_bin['equ_depth__departure_delay_in_minutes'].apply(
    lambda x: group_mapping[x])
```

```python
df_eq_bin['equ_depth__arrival_delay_in_minutes'] = pandas.qcut(
    df_eq_bin['Arrival Delay in Minutes'],
    8,  duplicates='drop'
)

# labeling
grouplist = df_eq_bin['equ_depth__arrival_delay_in_minutes'].unique(
).tolist()
grouplist.sort()
group_mapping = {g: "bin_{}".format(i) for i, g in enumerate(grouplist)}
df_eq_bin['equ_depth__arrival_delay_in_minutes'] = df_eq_bin['equ_depth__arrival_delay_in_minutes'].apply(
    lambda x: group_mapping[x])
```

## 2.Normalize with Min-Max and Z-Score

Flight Distance


```python
# min-max

# transform


def min_max_transform(list_set: pandas.Series) -> pandas.Series:
    new_set: pandas.Series = list_set.copy()
    new_set = (list_set - list_set.min()) / (list_set.max() - list_set.min())
    return new_set


def z_score_transform(list_set: pandas.Series) -> pandas.Series:
    new_set: pandas.Series = list_set.copy()
    new_set = (list_set - list_set.mean())/list_set.std()
    return new_set


df_minmax['Flight Distance'] = min_max_transform(df_copy['Flight Distance'])
df_zscore['Flight Distance'] = z_score_transform(df_copy['Flight Distance'])
```

```python

# df_flight_distance = df_copy[['Flight Distance']]
df_flight_distance = df_copy.filter(['Flight Distance'], axis=1)

df_flight_distance['flight_distance__min_max'] = min_max_transform(
    df_copy['Flight Distance'])
df_flight_distance['flight_distance__z_score'] = z_score_transform(
    df_copy['Flight Distance'])


df_flight_distance
```

Therefore, for following summary methorlogy may use Decision Trees 

```python
def decision_tree_rend(trainset: pandas.DataFrame, resultset: pandas.DataFrame, test_ratio: float):
    from sklearn import tree
    from sklearn.model_selection import train_test_split

    X_train, X_test, y_train, y_test = train_test_split(
        trainset, 
        resultset, 
        test_size=test_ratio)

    model = tree.DecisionTreeClassifier( max_depth=5)
    model = model.fit(X_train, y_train)
    _score = model.score(X_test, y_test)
    return model, _score 


```
